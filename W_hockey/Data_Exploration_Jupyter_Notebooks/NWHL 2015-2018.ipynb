```python
columns = NWHL_all_format_2015.columns.tolist()
column_slice = columns[6:]
column_slice
```

```python
for name in one_name:
    df = NWHL_all_format_2015[NWHL_all_format_2015['Player'] == name]
    player_stat_fin = [2015, 'NWHL']
    player_stat_col = NWHL_all_format_2015.columns[:6]
    player_stat = df.iloc[0][2:5]
    player_stat2 = player_stat.tolist()
    player_stat2.insert(0, 2015)
    player_stat2.insert(1, 'NWHL')
    player_stat2.append(df.shape[0])
    #Season_2015_agg.loc[0] = player_stat2
    for col in column_slice:
        val = df[col].sum()
        player_stat2.append(val)
    player_stat3 = pd.Series(player_stat2, index=columns)
    NWHL_all_format_2015 = NWHL_all_format_2015.append(player_stat3, ignore_index=True)
#Season_2015_agg.append(player_stat2, ignore_index=True)
#player_stat2
NWHL_all_format_2015.shape
```
